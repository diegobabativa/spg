### 1. Loading pre-processed data

```python
import pandas as pd

df = pd.read_parquet('../../data/proccessed/interview_dataset_processed.parquet')
df.shape
```

```python
categorical_variables = (
    list(df.drop(columns=['date_on_production'])
           .select_dtypes(exclude=['float','number'])
           .columns)
        )
categorical_variables
```

```python
df['treatment_company'] = df['treatment_company'].str.replace(' ', '_').str.replace('treatment_company', 'treat_comp')
df['operator'] = df['operator'].str.replace(' ', '_').str.replace('operator', 'oper')
```

```python
df[categorical_variables]
```

```python
df_dummies_cat = pd.get_dummies(
        df[categorical_variables],
        drop_first=True,
        prefix={'treatment_company':'tc',
                'operator':'opt',}
        )
df_dummies_cat
```

```python
df_features = (
    df.drop(columns=categorical_variables)
    .join(df_dummies_cat, how='inner')
)
```

```python
list(df_features.columns)
```

```python
df_features.isna().sum()
```

```python
df_features = df_features.drop(columns=['date_on_production'])
```

```python
df_features.to_parquet('../../data/features/interview_dataset_features_v1.parquet')
```
